```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

import scipy.stats as stats

from scipy.interpolate import spline
```

```python
mpl.rcParams['font.size'] = 16
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16 
```

```python
fig, ax = plt.subplots(1,2, figsize = (14, 6))
rv = stats.norm(loc = 2.5, scale = 1)
ru = stats.uniform(loc = 0, scale = 5)
x = np.linspace(0, 5)
ax[0].plot(x, ru.cdf(x), 'k--')
ax[0].set_xlabel(r'$X$: Uniform')
ax[0].set_ylabel(r'Cumulative distribution')

ax[1].plot(x, rv.cdf(x), 'k--')
ax[1].set_xlabel(r'$Y$: Normal')
ax[1].set_ylabel(r'Cumulative distribution')
plt.savefig('cdf_inverse.pdf', dpi=300)
```
